```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import xlrd
```

```python
open("Salthalt_CTD_Skagerak.csv")
salinitet = pd.read_csv('Salthalt_CTD_Skagerak.csv')
```

```python
lat  = salinitet.iloc[0,1:]
depth = salinitet.iloc[1:,0]
sal = salinitet.iloc[1:,1:]

print(np.size(lat))
print(np.size(depth))
print(np.size(sal))
```

```python
#print(lat)
#print(depth)
#print(sal)
```

```python
plt.contourf(lat,depth,sal,np.arange(15, 33, 0.1))
cbar = plt.colorbar()
#plt.contour(lat,depth,sal,np.arange(35, 37, 0.1),colors='red')
plt.gca().invert_yaxis()
plt.xlabel(r'Latitude ($^\circ$N)',fontsize=18)
plt.ylabel('Depth (m)',fontsize=16)
plt.title('Salinity transect',fontsize=20)
cbar.set_label('Salinity (PSU)',fontsize=16)
plt.show()
```
