```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('obama.csv', parse_dates=['year_month'])
data.year_month
```

```python
plt.plot(data.year_month, data.approve_percent, 'o', markersize=2,
         alpha=0.3)
plt.show()
```

```python
sampled = data.sample(frac=0.5)
# Use n=1000 to specify a number of samples instead.
plt.plot(sampled.year_month, sampled.approve_percent, 'o', markersize=2,
         alpha=0.3)
plt.show()
```
